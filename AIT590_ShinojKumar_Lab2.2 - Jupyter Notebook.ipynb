```python
#N gram modelfor Next Word Prediction.
# Data is taken from webpage and normalized.
# Model is based on Trigram and biGram *
import nltk
from nltk.tokenize import sent_tokenize
from nltk.tokenize import WordPunctTokenizer
from nltk import ngrams
from collections import Counter
import bs4 as bs  # BeautifulSoup
import urllib.request
import re
```

```python
#Using the web scraping technique with BeautifulSoup
def _scrape_webpage(url):
       
    scraped_textdata = urllib.request.urlopen(url)
    textdata = scraped_textdata.read()
    parsed_textdata = bs.BeautifulSoup(textdata,'lxml')
    paragraphs = parsed_textdata.find_all('p')
    formated_text = ""
    
    for para in paragraphs:
        formated_text += para.text
    
    return formated_text
```

```python
mytext = _scrape_webpage('https://en.wikipedia.org/wiki/Natural_language_processing')


modifiedText = []
punctuation = '''!"#$%&'()*+,-/:;<=>?@[\]^_`{|}~''' # '.' has been excluded from the List
```

```python
tokens = nltk.word_tokenize(mytext)

#Create your bigrams
bgs = nltk.bigrams(tokens)

#compute frequency distribution for all the bigrams in the text
fdist = nltk.FreqDist(bgs)
for k,v in fdist.items():
    print(k,v)
```

```python
m = []
for i in sent_tokenize( mytext ):
  for j in WordPunctTokenizer().tokenize(i):
    if j not in punctuation:
      m.append( j.lower() )

uniqueWords = set( m )
uniGramList = list( ngrams( m,1 ))
uniGramCounter = Counter(uniGramList)


biGramList = list( ngrams( m,2 ))
biGramCounter = Counter(biGramList)
    
triGramList = list( ngrams( m,3 ))
triGramCounter = Counter(triGramList)

#print( uniGramList )
print( uniGramCounter )
#print( biGramList )
print( biGramCounter )
#print( triGramList )
print( triGramCounter )
#print(uniqueWords)
```

```python
def findNextWord( word1, word2 ):
  #
  word1 = word1.lower()
  word2 = word2.lower()
  
  nextWordProbDic = {}
  biGramTupple = word1, word2
  biGramTuppleOccurance = biGramCounter[ biGramTupple ]
  print("biGramTupple: ",biGramTupple,",===> ",biGramTuppleOccurance )
  
  for i in uniqueWords:
    triGramTupple = word1,word2,i
    triGramTuppleOccurance = triGramCounter[ triGramTupple ]
    
    if triGramTuppleOccurance > 0:
      print("triGramTupple: ",triGramTupple,",===> ",triGramTuppleOccurance )
      probChance = ( triGramTuppleOccurance / biGramTuppleOccurance )
      
      try:
        listOb = nextWordProbDic[probChance]
        listOb.append( i )
        
      except KeyError:
        listOb = [i]
        nextWordProbDic[probChance] = listOb
        
  maxProb = max( nextWordProbDic.keys() )
  print("Among: ",nextWordProbDic,"\nSuggested word is :  ")
  for i in nextWordProbDic[ maxProb ]:
    print(i)
```
